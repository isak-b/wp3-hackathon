# LangChain basics

docs: [AzureChatOpenAI](https://python.langchain.com/api_reference/openai/chat_models/langchain_openai.chat_models.azure.AzureChatOpenAI.html)

Set your environment variables:

```bash
    export AZURE_OPENAI_API_KEY=<your api key>
    export AZURE_OPENAI_ENDPOINT=<your api endpoint>
```


````python
import os

import langchain_core
from langchain_openai import AzureChatOpenAI

# Load credentials from the environment
credentials = {
    "api_key": os.environ["AZURE_OPENAI_API_KEY"],
    "azure_endpoint": os.environ["AZURE_OPENAI_ENDPOINT"],
    "api_version": "2025-01-01-preview",
}

# Client & Model
model = "gpt-4o-mini"
client = AzureChatOpenAI(azure_deployment=model, **credentials)

# Input messages
messages = [
    langchain_core.messages.SystemMessage("You are a helpful assistant"),
    langchain_core.messages.HumanMessage("Hello"),
]

# Call the model
response = client.invoke(messages)
messages.append(response)

# Print messages
for message in messages:
    print(f"{type(message).__name__}: {message.content}")
````
