# LangChain basics

docs: [ChatOpenAI](https://docs.langchain.com/oss/python/integrations/chat/openai)

See [README.md](/README.md) for how to set up environment.


```python
import os
from dotenv import load_dotenv
import langchain.messages
from langchain_openai import ChatOpenAI

load_dotenv();

# Client & Model
client = ChatOpenAI(
    model="gpt-4o-mini",
    base_url=os.environ["OPENAI_BASE_URL"],
    api_key=lambda: os.environ["OPENAI_API_KEY"]
)

# Input messages
messages = [
    langchain.messages.SystemMessage("You are a helpful assistant"),
    langchain.messages.HumanMessage("Hello"),
]

# Call the model
response = client.invoke(messages)
messages.append(response)

# Print messages
for message in messages:
    print(f"{type(message).__name__}: {message.content}")
```
